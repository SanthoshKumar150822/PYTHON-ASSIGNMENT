```python
import pandas as pd
from matplotlib import pyplot as plt
iris=pd.read_csv('C:/Users/HOME/Downloads/Cricket7.csv')
setosa=iris[iris['NAME']=='M.S.DHONI']  #setosa has 50 records so 50 datapoints
setosa.plot(x="6S",y="4S",kind="scatter",label='setosa',color='b')
plt.title('scatter plot')
plt.xlabel('6S')
plt.ylabel('4S')
plt.show()
```

```python
setosa["100S"].plot(kind="line",label="dhoni",color='black',linewidth='1.5',ls='-') #ls-plot constructeor
plt.axis([0,500,0,100])
plt.title('Line Plot')
plt.xlabel('6S')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
iris=pd.read_csv('C:/Users/HOME/Downloads/Cricket7.csv')
Type=list(iris['ODI Runs'].unique())
count=list(iris['ODI Runs'].value_counts())
avg=list(iris.groupby('ODI Runs')["ODI Runs"].mean())
plt.bar(Type,avg,color=['maroon','yellow','blue'],width=0.4)

plt.title('Bar Plot')
plt.xlabel("Average ODI Runs")
plt.ylabel("ODI Runs")
plt.show()
```
